```python
import pandas as pd
import numpy as np
from apyori import apriori as apri
```

```python
df=pd.read_csv('movie_dataset.csv', header=None)
```

```python
df.head()
```

```python
df.shape
```

```python
records = []
for i in range(1, 7501):
 records.append([str(df.values[i, j]) for j in range(0, 20)])
```

```python
association_rules = apri(records, min_support=0.01, min_length=2)
association_results = list(association_rules)
association_results
```

```python
association_rules = apri(records, min_support=0.01, min_confidence=0.5, min_length=2)
association_results = list(association_rules)
association_results
```
